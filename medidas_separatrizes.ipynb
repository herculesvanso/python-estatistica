```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados  = pd.read_csv('dados.csv')
dados.head()
```

```python
# quartil
dados.Renda.quantile([.25,.50,.75])
```

```python
#decil
dados.Renda.quantile([ i / 10 for i in range(1,10)])
```

```python
# percentil
dados.Renda.quantile([ i / 100 for i in range(1,100)])
```

```python
sns.histplot(dados.Idade, cumulative = True)
plt.ylabel('Frequência')
```

```python
ax = sns.boxplot(x = 'Altura', data = dados, orient = 'h')
ax.figure.set_size_inches(12,6)
```

```python
quantil_75 = dados.Renda.quantile(.75)
quantil_25 = dados.Renda.quantile(.25)
iiq = quantil_75 - quantil_25

lim_inf = quantil_25 - 1.5 * iiq
lim_sup = quantil_75 + 1.5 * iiq

print(lim_inf)
print(lim_sup)
```

```python
mask = ((lim_inf <= dados.Renda) & (dados.Renda <= lim_sup))
```

```python
ax = sns.boxplot(x='Renda',data = dados[mask], orient = 'h')
ax.figure.set_size_inches(12,6)
```

```python
ax = sns.boxplot(x='Renda', y = 'Sexo',data = dados[mask], orient = 'h')
ax.figure.set_size_inches(12,6)
```
