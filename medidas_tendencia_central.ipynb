```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = pd.read_csv('dados.csv')
dados.head()
```

# Média

```python
dados.Renda.mean()
```

```python
dados.groupby('Sexo').mean()
```

```python
media_renda_sexo = dados.groupby('Sexo').mean()['Renda']
media_renda_sexo
```

```python
media_renda_sexo = pd.DataFrame(media_renda_sexo).round(2)
media_renda_sexo.rename(columns = {'Renda':'Renda Média'}, inplace = True)
media_renda_sexo.rename(index = {0:'Masculino',1:'Feminino'}, inplace = True)
media_renda_sexo
```

# Mediana

```python
mediana_renda_sexo = dados.groupby('Sexo').median()['Renda']
mediana_renda_sexo
```

```python
sexo = {1:'Masculino',2:'Feminino'}
```

```python
mediana_renda_sexo = pd.DataFrame(mediana_renda_sexo)
mediana_renda_sexo.rename(columns = {'Renda':'Renda Média'}, inplace = True)
mediana_renda_sexo.rename(index = sexo, inplace = True)
mediana_renda_sexo
```

# Moda

```python
moda_renda= dados.Renda.mode()
moda_renda
```

```python
dados[dados.Renda == moda_renda.values[0]].Renda.value_counts()
```

# Simetria

```python
ax = sns.histplot(dados.query('Renda < 10000').Renda, kde = True)
ax.figure.set_size_inches(12,6)
ax
```

```python
dados.Renda.mean()
```

```python
dados.Renda.median()
```

```python
dados.Renda.mode()
```
